```python
img1 = cv2.imread('img1.jpg')
img2 = cv2.imread('img2.jpg')
```

```python
img1 = cv2.resize(img1, (300, 300))
img2 = cv2.resize(img2, (300, 300))
```

```python
added_image = cv2.add(img1, img2)
subtracted_image = cv2.subtract(img1, img2)
```

```python

# Convert images to grayscale
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# Apply threshold to create binary masks
_, mask1 = cv2.threshold(gray1, 127, 255, cv2.THRESH_BINARY)
_, mask2 = cv2.threshold(gray2, 127, 255, cv2.THRESH_BINARY)

bitwise_and = cv2.bitwise_and(img1, img2, mask=mask1)
bitwise_or = cv2.bitwise_or(img1, img2, mask=mask1)
bitwise_xor = cv2.bitwise_xor(img1, img2, mask=mask1)
bitwise_not = cv2.bitwise_not(img1, mask=mask1)

```

```python
# Convert BGR (OpenCV format) to RGB (Matplotlib format)
def show_img(title, img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

show_img("Image 1", img1)
show_img("Image 2", img2)
show_img("Added Image", added_image)
show_img("Subtracted Image", subtracted_image)
show_img("Bitwise AND", bitwise_and)
show_img("Bitwise OR", bitwise_or)
show_img("Bitwise XOR", bitwise_xor)
show_img("Bitwise NOT (Image 1)", bitwise_not)
```
